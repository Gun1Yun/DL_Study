```python
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00466/'
data_name = 'CNFUV_Datasets.zip'
r = requests.get(data_url+data_name)
files = zipfile.ZipFile(io.BytesIO(r.content))
lst = files.namelist()
f = lst[1:5]+lst[6:]
df = pd.concat(
    [pd.read_excel(files.open(i)) for i in f]
)
df
```
